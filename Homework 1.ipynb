#### Q:3 Meaning of each column
1. **Player** = Player's Name
2. **Span** = How many years they have played cricket professionally
3. **Mat** = Total number of cricket matches they participated in
4. **Inns** = Total number of innings they played
5. **Balls** = Total number of balls they played
6. **Runs** = Total number of runs they gave away
7. **Wkts** = Total number of wickets they took
8. **Ave** = Average, Ratio of number of runs scored to the number of wickets taken by the bowler
9. **Econ** = Economy, average number of runs they have conceded per over
10. **SR** = Strike Rate, the average number of balls bowled per wicket taken
11. **4** = Count of 4s/boundaries batsmen scored from their balls
12. **5** = Number of five-wicket hauls they had

#### Q8. How many players played for ICC?

```python
#To split Player column and create two new columns : splitting the 'Player' column to get the information about 'Country'

df_player = df['Player'].str.split("(", expand=True)


#Adding Player_Name and Country columns in the data set

df[["Player_Name", "Country"]] = df['Player'].str.split("(", expand=True)

#Removing Player column
df = df.drop('Player', axis=1)

#Removing ')' value from Country column

df['Country'] = df['Country'].str.replace(")", "")
display(df.head(10))
```

#### Q9. How many different countries are present in this dataset? 

```python
#Unique Country list
df['Country'].unique()

#Value Counts of Country Column
df['Country'].value_counts()

#Checking where ICC, Asia, Africa have or not
def icc_check(i):
    if "ICC" in i:
        return "Yes"
    else:
        return "No"
    
def asia_check(x):
    if "Asia" in x:
        return "Yes"
    else:
        return "No"

def africa_check(x):
    if "Afr" in x:
        return "Yes"
    else:
        return "No"
    
df['Played_for_ICC'] = df['Country'].apply(icc_check)
df['Played_for_Asia'] = df['Country'].apply(asia_check)
df['Played_for_Africa'] = df['Country'].apply(africa_check)

#Finding how many players played for ICC
print('ICC:',df['Played_for_ICC'].value_counts()) 


#Removing ICC/ /ICC Asia/ Afr/ from the 'Country'
df['Country'] = df['Country'].str.replace("ICC/", "")
df['Country'] = df['Country'].str.replace("/ICC", "")
df['Country'] = df['Country'].str.replace("Asia/", "")
df['Country'] = df['Country'].str.replace("Afr/", "")

display(df.head(10))

#Checking the unique values for Country column
df['Country'].value_counts()
```

There are **11 different countries** are present in this dataset.

#### Q10. Which player(s) had played for the longest period of time?

```python
# splitting the 'Playing span' column based on the "-"
df[['Start_year', 'End_year']] = df['Playing span'].str.split("-", expand=True)

# removing the "Playing span" column
df = df.drop("Playing span", axis=1)

#Data type conversion
df['Start_year'] = df['Start_year'].astype('int') 
df['End_year'] = df['End_year'].astype('int')
print(df.info())

#Finding the years played of each player
df['Years_played'] = df['End_year'] - df['Start_year'] + 1

#Removing Start_year and End_year from data set
df = df.drop(['Start_year', "End_year"], axis=1) 

display(df.head(10))
```

```python
#Players had played for the longest period of time
df.sort_values(by='Years_played', ascending = False).head(5)
```

SR Tendulkar had played for the **longest period** of time.
